```python

from ultralytics import  YOLO
import cv2 as cv
import threading

class Mywebcam:
    def __init__(self,model_path):
        self.cam=cv.VideoCapture(0) #
        if self.cam.isOpened()==False:
            print("open camera failure")
            exit(0)
        self.cam.set(cv.CAP_PROP_FRAME_WIDTH, 1280) #
        self.cam.set(cv.CAP_PROP_FRAME_HEIGHT, 640)
        self.model=YOLO(model_path)    #model_path
        self.success,self.frame=self.cam.read() #
        if self.success == 0:
            print("read camera failure")
            exit(0)
        self.dealed_frame=self.model(self.frame)[0].plot()
        
        thread1=threading.Thread(target=self.read)#
        thread1.start()
        thread2=threading.Thread(target=self.deal)
        thread2.start()
    def read(self):
        while 1:
            self.suceess,tem=self.cam.read()
            if self.success:
                self.frame=tem
            else :
                print("read camera failure")
                exit(0)
        
    def deal(self):
        while 1:
            self.dealed_frame=self.model(self.frame)[0].plot()
        
        
        

if __name__=='__main__':
    title = "YOLOv8 Inference"
    mywebcam=Mywebcam(r"./models/ultimate_model/weights/best.pt")
    while 1:
        if mywebcam.dealed_frame is None:
            print("None")
            continue
        cv.imshow(title,mywebcam.dealed_frame)
        cv.waitKey(20)
    
    
    
    

```

```python
import cv2

from ultralytics import YOLO

#  YOLOv8 
model = YOLO(r"./models/ultimate_model/weights/best.pt")

# 
# cap = cv2.VideoCapture("path/to/your/video/file.mp4")
# 0
cap = cv2.VideoCapture(0)

# 
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 640)

title = "YOLOv8 Inference"
# 
cv2.namedWindow(title, cv2.WINDOW_NORMAL)
cv2.moveWindow(title, 1280, 200)

# 
while cap.isOpened():
    # 
    success, frame = cap.read()
    if success:
        #  YOLOv8 
        results = model(frame)
        # 
        annotated_frame = results[0].plot()
        # 
        cv2.imshow(title, annotated_frame)
        # q
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # 
        break

# 
cap.release()
cv2.destroyAllWindows()
```
